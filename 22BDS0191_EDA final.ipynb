```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("TravelMode.csv")
df.head()
```

```python
# Select a numerical column for analysis
num_col = "income"
# Calculate summary statistics using the 'income' column from TravelMode.csv
mean_val = df[num_col].mean()
median_val = df[num_col].median()
# .mode() can return multiple values if they have the same highest frequency; .values[0] takes the first one.
mode_val = df[num_col].mode().values[0]
skewness_val = df[num_col].skew()
kurtosis_val = df[num_col].kurt()
# Print summary statistics
print("Summary Statistics for 'income':")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Mode: {mode_val:.2f}")
print(f"Skewness: {skewness_val:.2f}")
print(f"Kurtosis: {kurtosis_val:.2f}")
```

```python
# Create a skewness plot (using histogram) for 'income'
plt.figure(figsize=(10, 6))
sns.histplot(df[num_col], kde=False)
plt.title("Skewness Plot (Histogram) for 'income'")
plt.xlabel(num_col)
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a kurtosis plot (using kernel density estimate) for 'income'
plt.figure(figsize=(10, 6))
sns.kdeplot(df[num_col], fill=True)
plt.title("Kurtosis Plot (KDE) for 'income'")
plt.xlabel(num_col)
plt.ylabel("Density")
plt.show()
```

```python
# Create a box plot for 'income'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df[num_col])
plt.title("Box Plot for 'income'")
plt.xlabel(num_col)
plt.show()
```

```python
X_cluster = df.iloc[:, [6, 8]].values # Use a distinct variable name

## Using the elbow method to find the optimal number of clusters
# from sklearn.cluster import KMeans # Already imported
wcss = []
for i in range(1, 11):
    # Using n_init='auto' to suppress future warnings and potentially get better results
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42, n_init='auto')
    kmeans.fit(X_cluster)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10, 6)) # Added figure size for better plot readability
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method (Travel Time vs Income)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Training the K-Means model on the dataset
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42, n_init='auto')
y_kmeans = kmeans.fit_predict(X_cluster)

## Visualising the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X_cluster[y_kmeans == 0, 0], X_cluster[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_cluster[y_kmeans == 1, 0], X_cluster[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_cluster[y_kmeans == 2, 0], X_cluster[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X_cluster[y_kmeans == 3, 0], X_cluster[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of travelers (Travel Time vs Income)')
plt.xlabel('Travel Time')
plt.ylabel('Income')
plt.legend()
plt.show()
```

```python
#Dendograms
X_dendro = df.iloc[:,[6,8]].values
plt.figure(figsize=(15,6))
plt.title('Dendrogram (Travel Time vs Income)')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
#plt.grid(True)
dendrogram = sch.dendrogram(sch.linkage(X_dendro, method = 'ward'))
plt.show()
```

```python
# Preparing the data for Linear Regression
df_encoded = pd.get_dummies(df, columns=['mode'], drop_first=True)
df_encoded.head()
```

```python
# 2. Define features (X) and target (y)
X = df_encoded.drop(['rownames', 'individual', 'choice', 'gcost'], axis=1)
y = df_encoded['gcost']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
# 4. Training the Multiple Linear Regression Model
multi_regressor = LinearRegression()
# Fitting the training data to our model
multi_regressor.fit(X_train, y_train)
```

```python
# 5. Model Evaluation for Multiple Linear Regression
# check prediction score/accuracy on test data
print(f"\nMultiple Linear Regression R^2 score on test data: {multi_regressor.score(X_test, y_test)}")
```

```python
# predict the y values
y_pred = multi_regressor.predict(X_test)
# a data frame with actual and predicted values of y
eval = pd.DataFrame({'Actual': y_test.values.flatten(),
                               'Predicted': y_pred.flatten()})
print("\nActual vs Predicted values (Multiple Regression - first 10):")
print(eval.head(10))
```

```python
# Plotting Actual vs Predicted for Multiple Regression (first 10)
eval.head(10).plot(kind = 'bar', figsize=(10, 6))
plt.title('Actual vs Predicted General Cost (Multiple Regression - First 10 Test Samples)')
plt.ylabel('General Cost')
plt.xlabel('Sample Index')
plt.show()
```

```python


# Computing accuracy (metrics) for Multiple Linear Regression
# Scoring the model
print("\nMultiple Linear Regression Model Evaluation Metrics:")
# R2 Score
print(f"R2 score: {r2_score(y_test, y_pred)}")
# Mean Absolute Error (MAE)
print(f"MAE score: {mean_absolute_error(y_test, y_pred)}")
# Mean Squared Error (MSE)
print(f"MSE score: {mean_squared_error(y_test, y_pred)}")
# Root Mean Squared Error (RMSE)
print(f"RMSE score: {np.sqrt(mean_squared_error(y_test, y_pred))}")
```
